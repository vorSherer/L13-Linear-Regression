### SETUP

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
ic_df = pd.read_csv('./data/datasets_561327_1020381_IceCreamData.csv')
ic_df.head()
```

#### Dataset description
__About this file__ - This is a simple sample dataset for training.

"You own an ice cream business and you would like to create a model that could predict the daily revenue in dollars based on the outside air temperature (degC)You own an ice cream business and you would like to create a model that could predict the daily revenue in dollars based on the outside air temperature (degC)"

```python
X = ic_df['Temperature'].values
y = ic_df['Revenue'].values
```

```python
X = X.reshape(-1,1)
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=100)
```

```python
print(f"X_train shape = {X_train.shape}.")
print(f"y_train shape = {y_train.shape}.")
print(f"X_test shape = {X_test.shape}.")
print(f"y_test shape = {y_test.shape}.")
```

```python
plt.scatter(X_train, y_train, color='red')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.title('Ice Cream sales revenue as a function of temperature')
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_predict = lr.predict(X_test)
print(f"Train accuracy: {round(lr.score(X_train, y_train) * 100,2)} %")
print(f"Test accuracy: {round(lr.score(X_test, y_test) * 100,2)} %")
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_test, y_predict)
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.title('Ice Cream sales revenue as a function of temperature')
plt.show()
```

## Conclusions:
1. In this dataset, Ice Cream sales revenues directly corresponded to ambient temperature with a high degree of confidence.
1. Though randomly selected, data in the train and test sets were very similar in plot.
1. A broader dataset (e.g., including other weather factors and their impact on sales revenues) might have provided a more informative window into projecting revenues.
